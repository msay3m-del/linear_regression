# Linear Regression Analysis (Python)

## Project Objective

This notebook fits a simple linear regression model to predict employee **Salary** from **YearsExperience**. We compute regression diagnostics (slope, intercept, correlation, MSE, and R-squared) and visualize the relationship with a scatter plot and fitted regression line.

## Data Loading

We load the dataset from `regression_data.csv` using pandas. Each row represents an employee with years of experience and corresponding salary.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv("regression_data.csv")
print(f"Loaded {len(data)} observations.")
data.head()
```

## Model Fitting

We fit a linear regression model with **YearsExperience** as the predictor and **Salary** as the response variable using scikit-learn's `LinearRegression`.

```python
x = data["YearsExperience"]
y = data["Salary"]

x_values = x.values.reshape(-1, 1)
y_values = y.values

model = LinearRegression()
model.fit(x_values, y_values)
predictions = model.predict(x_values)

slope = model.coef_[0]
intercept = model.intercept_
r_value, _ = pearsonr(x, y)
mse = mean_squared_error(y_values, predictions)
r_squared = r2_score(y_values, predictions)

print("Linear Regression Results")
print("=" * 40)
print(f"Slope:                 {slope:.4f}")
print(f"Intercept:             {intercept:.4f}")
print(f"Pearson r:             {r_value:.4f}")
print(f"Mean Squared Error:    {mse:.4f}")
print(f"R-squared:             {r_squared:.4f}")
print(f"Equation:              y = {slope:.4f}x + {intercept:.4f}")
```

## Slope Interpretation

The slope represents the average change in salary for each additional year of experience. A positive slope indicates that salary tends to increase as experience increases.

## Intercept Interpretation

The intercept is the predicted salary when years of experience equals zero. In practice, this value may be a theoretical baseline rather than a realistic salary for new hires.

## Correlation Interpretation

The Pearson correlation coefficient (r) measures the strength and direction of the linear relationship between years of experience and salary. Values close to +1 indicate a strong positive linear association.

## MSE Interpretation

Mean Squared Error (MSE) is the average of squared differences between observed and predicted salaries. Lower MSE values indicate predictions that are closer to the actual data on average.

## Visualization

The scatter plot shows individual observations, and the regression line represents the fitted model. Annotations display the regression equation and correlation coefficient.

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.scatter(x, y, alpha=0.7, edgecolors="black", linewidth=0.5, label="Observations")
ax.plot(x, predictions, color="red", linewidth=2, label="Regression line")

equation_text = f"y = {slope:.2f}x + {intercept:.2f}"
correlation_text = f"r = {r_value:.4f}"

ax.text(
    0.05,
    0.95,
    f"{equation_text}\n{correlation_text}",
    transform=ax.transAxes,
    fontsize=12,
    verticalalignment="top",
    bbox=dict(boxstyle="round", facecolor="white", alpha=0.85),
)

ax.set_xlabel("Years of Experience")
ax.set_ylabel("Salary")
ax.set_title("Linear Regression: Salary vs Years of Experience")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.4)

plt.tight_layout()
plt.savefig("regression_plot_python.png", dpi=150)
plt.show()
```

## Conclusions

The linear model captures the general upward trend between experience and salary. The slope quantifies expected salary growth per year, while R-squared and MSE summarize model fit. Residual patterns in the plot can help assess whether a simple linear model is appropriate for this dataset.
