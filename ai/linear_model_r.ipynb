# Linear Regression Analysis (R)

## Project Objective

This notebook fits a simple linear regression model to predict employee **Salary** from **YearsExperience**. We compute regression diagnostics (slope, intercept, correlation, MSE, and R-squared) and visualize the relationship with a scatter plot and fitted regression line.

## Data Loading

We load the dataset from `regression_data.csv`. Each row contains years of experience and the corresponding salary.

```r
library(ggplot2)

data <- read.csv("regression_data.csv", stringsAsFactors = FALSE)
cat(sprintf("Loaded %d observations.\n", nrow(data)))
head(data)
```

## Model Fitting

We fit a linear model using R's `lm()` function with **Salary** as the response and **YearsExperience** as the predictor.

```r
model <- lm(Salary ~ YearsExperience, data = data)

slope <- coef(model)["YearsExperience"]
intercept <- coef(model)["(Intercept)"]
predictions <- fitted(model)

r_value <- cor(data$YearsExperience, data$Salary, method = "pearson")
mse <- mean((data$Salary - predictions)^2)
r_squared <- summary(model)$r.squared

cat("Linear Regression Results\n")
cat(paste(rep("=", 40), collapse = ""), "\n")
cat(sprintf("Slope:                 %.4f\n", slope))
cat(sprintf("Intercept:             %.4f\n", intercept))
cat(sprintf("Pearson r:             %.4f\n", r_value))
cat(sprintf("Mean Squared Error:    %.4f\n", mse))
cat(sprintf("R-squared:             %.4f\n", r_squared))
cat(sprintf("Equation:              y = %.4f x + %.4f\n", slope, intercept))
```

## Slope Interpretation

The slope represents the average change in salary for each additional year of experience. A positive slope indicates that salary tends to increase as experience increases.

## Intercept Interpretation

The intercept is the predicted salary when years of experience equals zero. It serves as a baseline estimate and may not reflect a realistic salary for employees with no experience.

## Correlation Interpretation

The Pearson correlation coefficient (r) measures the strength and direction of the linear relationship between years of experience and salary. Values near +1 indicate a strong positive linear association.

## MSE Interpretation

Mean Squared Error (MSE) is the average squared difference between observed and predicted salaries. Lower MSE values indicate better predictive accuracy on average.

## Visualization

The scatter plot displays individual observations, and the regression line shows the fitted model. Annotations include the regression equation and correlation coefficient.

```r
equation_text <- sprintf("y = %.2f x + %.2f", slope, intercept)
correlation_text <- sprintf("r = %.4f", r_value)
annotation_text <- paste(equation_text, correlation_text, sep = "\n")

p <- ggplot(data, aes(x = YearsExperience, y = Salary)) +
  geom_point(alpha = 0.7, color = "steelblue", size = 3) +
  geom_smooth(method = "lm", se = FALSE, color = "red", linewidth = 1) +
  annotate(
    "label",
    x = min(data$YearsExperience) + 0.05 * diff(range(data$YearsExperience)),
    y = max(data$Salary),
    label = annotation_text,
    hjust = 0,
    vjust = 1,
    fill = "white",
    alpha = 0.85,
    size = 4
  ) +
  labs(
    title = "Linear Regression: Salary vs Years of Experience",
    x = "Years of Experience",
    y = "Salary"
  ) +
  theme_minimal()

print(p)
ggsave("regression_plot_r.png", plot = p, width = 10, height = 6, dpi = 150)
```

## Conclusions

The fitted linear model describes a positive relationship between experience and salary. Slope, R-squared, and MSE together summarize both the direction of the trend and the quality of the fit. Visual inspection of the scatter plot helps assess whether additional modeling complexity may be warranted.
